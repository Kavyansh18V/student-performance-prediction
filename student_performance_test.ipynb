```python
df_cleaned=df.copy()
df_cleaned
```

```python
df_cleaned.drop_duplicates(inplace=True)
```

```python
df_cleaned.shape
```

```python
df_cleaned.columns
```

```python
df_cleaned["gender"].value_counts()
```

```python
df_cleaned["gender"]=df_cleaned["gender"].map({"Male":0,"Female":1})
df_cleaned["internet_access"]=df_cleaned["internet_access"].map({"No":0,"Yes":1})
df_cleaned["extra_classes"]=df_cleaned["extra_classes"].map({"No":0,"Yes":1})

df_cleaned.rename(columns={
    'gender':'is_female',
})
```

```python
cat_columns=['parent_education']

df_cleaned=pd.get_dummies(df_cleaned,columns=cat_columns,drop_first=True,dtype=int)
df_cleaned
```

```python
df_cleaned.head()
```

```python
sns.countplot(df_cleaned,x="grade")
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

col_scaler=['student_id','study_hours_per_day','attendance_percentage','assignment_score', 'midterm_score', 'final_exam_score','participation_score', 'sleep_hours', 'overall_score']

df_cleaned[col_scaler]=scaler.fit_transform(df_cleaned[col_scaler])
df_cleaned.head()
```

```python
df_cleaned.columns
```

```python
from scipy.stats import pearsonr   #chexk relATION only for numerical

selected_features=['student_id', 'gender', 'study_hours_per_day', 'attendance_percentage',
       'assignment_score', 'midterm_score', 'final_exam_score',
       'participation_score', 'internet_access', 'extra_classes',
       'sleep_hours', 'overall_score', 'parent_education_High School',
       'parent_education_Master', 'parent_education_PhD']

correlations={
    features:pearsonr(df_cleaned[features],df_cleaned['final_exam_score'])[0]
    for features in selected_features
}

correlation_df = pd.DataFrame(list(correlations.items()), columns=['features', 'Pearson Correlation'])
correlation_df.sort_values(by='Pearson Correlation', ascending=False)
```

```python
from scipy.stats import chi2_contingency

cat_features=['gender', 'parent_education_High School',
       'parent_education_Master', 'parent_education_PhD', 'internet_access', 'extra_classes']
alpha = 0.05

chi2_results = {}

for col in cat_features:
    contingency = pd.crosstab(df_cleaned[col], df_cleaned['grade'])
    chi2_stat, p_val, _, _ = chi2_contingency(contingency)
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'Decision': decision
    }

chi2_df = pd.DataFrame(chi2_results).T
chi2_df = chi2_df.sort_values(by='p_value')
chi2_df
```

```python
df_final=df_cleaned.copy()
df_final
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df_final.drop("grade",axis=1)
y=df_final["grade"]
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report
acc=accuracy_score(y_test,y_pred)
print(classification_report(y_test,y_pred))


```

```python
print("Train accuracy:", model.score(X_train, y_train))
print("Test accuracy:", model.score(X_test, y_test))
```

```python
params = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga','lbfgs'],
}
```

```python
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(model,params,cv=5,scoring='accuracy')
grid.fit(X_train,y_train)
```
